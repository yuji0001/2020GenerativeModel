## Gumbel Softmax distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
u = np.random.rand(N)
plt.hist(u,bins = 100)
plt.show()
```

```python
pi = np.log(np.random.rand(N))
plt.hist(pi,bins = 100)
plt.show()
```

```python
g = -np.log(-np.log(u))
plt.hist(g,bins = 100)
plt.show()
```

```python
plt.hist(g+u,bins = 100)
plt.show()
```
